###End-to-End Pipeline for an Online Store 
####(Using Azure Databricks, Medallion Architecture with Unity Catalog, Azure SQL Database)

#### EXTERNAL LOCATION CREATION

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS Ext_Loc_Ecommerce
URL 'abfss://ecommerce-data@staccecommerce.dfs.core.windows.net/'
WITH (storage credential `storage-credential-ecommerce`);
```

####CATALOG CREATION

```python
%sql
CREATE CATALOG IF NOT EXISTS Ecommerce_Catalog;
```

####SCHEMA CREATION

```python
%sql
USE CATALOG Ecommerce_Catalog;

CREATE SCHEMA IF NOT EXISTS Bronze
MANAGED LOCATION 'abfss://ecommerce-data@staccecommerce.dfs.core.windows.net/Bronze/';

CREATE SCHEMA IF NOT EXISTS Silver
MANAGED LOCATION 'abfss://ecommerce-data@staccecommerce.dfs.core.windows.net/Silver/';

CREATE SCHEMA IF NOT EXISTS Gold
MANAGED LOCATION 'abfss://ecommerce-data@staccecommerce.dfs.core.windows.net/Gold/';
```
