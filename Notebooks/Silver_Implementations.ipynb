```python
from pyspark.sql.functions import explode, col

df_silver_carts = carts_df \
    .withColumn("cart", explode("carts")) \
    .select(
        col("cart.id").alias("cart_id"),          
        col("cart.userId").alias("userId"),
        col("cart.total").alias("cart_total"),
        col("cart.discountedTotal").alias("discountedTotal"),
        col("cart.totalProducts").alias("totalProducts"),
        col("cart.totalQuantity").alias("totalQuantity"),
        col("cart.products").alias("product_struct")  
    ) \
    .withColumn("product", explode("product_struct")) \
    .select(
        "*",
        col("product.id").alias("product_id"),  
        col("product.price").alias("price"),
        col("product.quantity").alias("quantity"),
        col("product.thumbnail").alias("thumbnail"),
        col("product.title").alias("product_title"),
        col("product.discountPercentage").alias("ProductdiscountPercentage"),
        col("discountedTotal").alias("ProductdiscountedTotal"),
        col("product.total").alias("product_total")
    ) \
    .drop("product", "product_struct")

display(df_silver_carts)
```

####TRANSFORMATIONS

```python
from pyspark.sql.functions import col, trim, lower, lit

def clean_silver_carts(df_silver_carts):
    """
    Cleans the Silver carts DataFrame
    - Drops rows with null essential columns
    - Casts types
    - Trims text
    - Filters invalid price/quantity
    """ 
    
    df_silver_carts_clean = df_silver_carts \
        .dropna(subset=["cart_id", "product_id", "product_title", "price"]) \
        .withColumn("order_date", to_date(lit(order_date), "yyyy-MM-dd"))\
        .withColumn("cart_id", col("cart_id").cast("int")) \
        .withColumn("userId", col("userId").cast("int")) \
        .withColumn("product_id", col("product_id").cast("int")) \
        .withColumn("quantity", col("quantity").cast("int")) \
        .withColumn("cart_total", col("cart_total").cast("double")) \
        .withColumn("discountedTotal", col("discountedTotal").cast("double")) \
        .withColumn("product_title", trim(col("product_title"))) \
        .filter((col("price") > 0) & (col("quantity") > 0))
    
    logger.info("df_silver_carts_clean created successfully.")
    display(df_silver_carts_clean)
    return df_silver_carts_clean
```

```python
df_silver_carts_clean = clean_silver_carts(df_silver_carts)
```

```python

from pyspark.sql.functions import sum

df_carts_agg = df_silver_carts_clean.groupBy(
    "product_id"
).agg(
    sum("quantity").alias("total_quantity")
)
```

```python
df_carts_agg.display()
```

```python

def clean_silver_products(df_silver_products):
    """
    Cleans the Silver products DataFrame
    - Drops rows with null product_id
    - Casts numeric columns
    - Trims and lowercases text columns
    - Filters invalid price and stock
    - Fills missing brand values
    - Drops unnecessary review columns
    """
    try:
        df_silver_products_clean = df_silver_products \
            .dropna(subset=["product_id"]) \
            .withColumn("order_date", to_date(lit(order_date)))\
            .withColumn("product_id", col("product_id").cast("int")) \
            .withColumn("price", col("price").cast("double")) \
            .withColumn("product_rating", col("product_rating").cast("double")) \
            .withColumn("stock", col("stock").cast("int")) \
            .withColumn("title", trim(col("title"))) \
            .withColumn("brand", lower(trim(col("brand")))) \
            .withColumn("category", lower(trim(col("category")))) \
            .filter((col("price") > 0) & (col("stock") >= 0)) \
            .fillna({"brand": "Unknown"}) \
            .drop("review_rating", "comment", "reviewerName", "reviewerEmail")
        
        logger.info("df_silver_products_clean created successfully.")
        display(df_silver_products_clean)
        return df_silver_products_clean

    except:
        logger.error(f"Error cleaning df_silver_products")
```

```python
df_silver_products_clean = clean_silver_products(df_silver_products)
```

```python
df_joined = df_carts_agg.join(
    df_silver_products_clean,
    on="product_id",
    how="inner"
)
```

```python
df_joined = df_joined.dropDuplicates(["product_id"])
```

####INCREMENTAL LOAD

```python
#STEP 1: Define table name and path

table_name = "ecommerce_catalog.silver.cleaned_data"
path = "abfss://ecommerce-data@staccecommerce.dfs.core.windows.net/Silver/cleaned_data"
```

```python
#STEP 2: Check if table exists

table_exists = spark.catalog.tableExists(table_name)
```

```python
#STEP 3: If table does NOT exist → First load

if not table_exists:
    df_joined.write.format("delta") \
        .mode("overwrite") \
        .option("path", path) \
        .saveAsTable(table_name)

    print("Initial load completed")
```

```python
#STEP 4: If table exists → Get last loaded date

from pyspark.sql.functions import max

if table_exists:
    last_date = spark.table(table_name) \
        .agg(max("order_date").alias("max_order_date")) \
        .collect()[0]["max_order_date"]
```

```python
# STEP 5: Filter only new data

from pyspark.sql.functions import col

if table_exists:
    df_new = df_joined.filter(col("order_date") > last_date)
```

```python
# Step 6: Incremental load using MERGE

from delta.tables import DeltaTable

if table_exists:
    delta_table = DeltaTable.forName(spark, table_name)

    delta_table.alias("target").merge(
        df_new.alias("source"),
        "target.product_id = source.product_id AND target.order_date = source.order_date"
    ).whenNotMatchedInsertAll().execute()

    print("Incremental load completed")
```
